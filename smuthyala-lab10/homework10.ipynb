```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
worker_stips = pd.read_csv("workerstips.csv")
flights_data = pd.read_csv("flightsData.csv")
titanic = pd.read_csv("titanic.csv")
```

```python
days = ['Thur', 'Fri', 'Sat', 'Sun']
worker_stips['day'] = pd.Categorical(worker_stips['day'], categories=days, ordered=True)
worker_stips = worker_stips.sort_values('day')
```

```python
sns.scatterplot(x="total_bill", y="tip", data=worker_stips)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=worker_stips)
```

```python
sns.barplot(x="day",y="tip", data=worker_stips)
```

## According to the Bar Plot, Sunday has the highest average tip.

```python
sns.boxplot(x="day", y="tip", hue="time",data=worker_stips)
```

```python
sns.lineplot(x="year", y="passengers", data=flights_data)
```

```python
year_data = flights_data.groupby('year').sum().reset_index()
sns.lineplot( x="year", y="passengers", data=year_data)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sns.countplot(x = 'Pclass', hue = 'Sex' ,data = titanic)
```
