```python
import pandas as pd
import numpy as np
```

```python
def main():
    ## a
    #calling the pd.Series with the list.
    series1=pd.Series([7,11,13,17]) 
    print('Series: a\n',series1)
    print('\n')

    ## b
    series2=pd.Series([100.0,100.0,100.0,100.0,100.0])
    print('Series: b\n',series2)

    ## c
    # 20 random integers between 0 and 100.
    rand_Int=np.random.randint(low=0,high=100,size=20)
    series3=pd.Series(rand_Int)
    print('\n')

    # describe returns statistic summary like mean ,std,etc
    stats=series3.describe() 
    print('Series: c\n',stats)
    print('\n')

    ## d
    # instead of default index 0,1,2,3 will have indices Julie,Charlie,Sam & Andrea. 
    temperatures=pd.Series([98.6,98.9,100.2,97.9],index=['Julie','Charlie','Sam','Andrea'])
    print('Series: d\n', temperatures)
    print('\n')


    ## e
    #dictionay key with names and values from temperatures
    series_d={'Julie':98.6,'Charlie':98.9,'Sam':100.2,'Andrea':97.9} 
    series5=pd.Series(series_d) 
    print('Series: e\n', series5)

main()
```

```python
# DataFrames
def dataFrames():

    ## f
    # Dictionary with key names and values as list of temperatures 
    dict_f={'Maxine':[56.4,23.6,34.7],'James':[5,6,77.8],'Amanda':[56.4,34.5,23.5]} 
    temp=pd.DataFrame(dict_f) 
    print('Data Frames: f\n', temp)
    print('\n')


    ## g
    dict_g={'Maxine':[96,43.8,32.8],'James':[34,56,77.8],'Amanda':[43.6,47.8,87.9]} 
    temp=pd.DataFrame(dict_g,index=['Morning','Afternoon','Evening']) 
    print('Data Frames: g\n', temp)
    print('\n')


    ## h
    maxine = temp['Maxine']
    print('Data Frames: h\n', maxine)
    print('\n')


    ## i
    #morning temperature 
    morning=temp.loc['Morning']
    print('Data Frames: i\n', morning)
    print('\n')

    ## j
    mor_eve=temp.loc[['Morning','Evening']] 
    print('Data Frames: j\n', mor_eve)
    print('\n')

    ## k
    #using list
    ama_max=temp[['Amanda','Maxine']]
    print('Data Frames: k\n', ama_max)
    print('\n')

    ## l
    row_col=temp.loc[['Morning','Afternoon'],['Amanda','Maxine']]
    print('Data Frames: l\n', row_col)
    print('\n')

    ## m
    stats_m=temp.describe()
    print('Data Frames: m\n', stats_m)
    print('\n')

    ## n
    # Transpose ,indices will become columns and columns will become indices
    temp_t=temp.T 
    print('Data Frames: n\n',temp_t)
    print('\n')

    ## o
    sort=temp.sort_index(axis=1) 
    print('Data Frames: o\n',sort)
    
dataFrames()
```
