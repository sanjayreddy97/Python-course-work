```python
#import libraries
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
def lineGraph():
    #file open and set mode as read
    profit = open("last_ten_year_net_profit.txt", "r")
    line = profit.readline() #reads firstline
    line = profit.readline() #reads secondline
    x_coords = []
    y_coords = []
    fig, ax = plt.subplots()
    while line != '':
        splitline = line.split(';')
        #remove spaces and appends to list
        x_coords.append(int(splitline[0].strip())) 
        #remove spaces & converting string to number
        y_coords.append(int(splitline[1].strip().replace('$','').replace(',',''))) 
        # reads nextline
        line = profit.readline()
    ax.plot(x_coords,y_coords)
    scale_y = 1e8
    plt.xticks(x_coords)
    #set y-axis label
    ax.set_ylabel('Net profit in the 100 millions')
    #set x-axis label
    ax.set_xlabel('years')
    #set title
    plt.title("Company's Profit Over the Past Ten Years")
    plt.grid()

lineGraph()
            
```

```python
def barGraph():
    profit = open("last_ten_year_net_profit.txt", "r")
    line = profit.readline()
    line = profit.readline()
    x_coords = []
    y_coords = []
    fig, ax = plt.subplots()
    while line != '':
        splitline = line.split(';')
        #remove spaces and appends to list
        x_coords.append(int(splitline[0].strip()))
        #remove spaces & converting string to number
        y_coords.append(int(splitline[1].strip().replace('$','').replace(',','')))
        #reads nextline
        line = profit.readline()
    #scale y-axis to 
    scale_y = 1e8
    plt.xticks(x_coords)
    #ticker to format y-axis values
    ticks_y = ticker.FuncFormatter(lambda x, pos: x/scale_y)
    ax.yaxis.set_major_formatter(ticks_y)
    #y-axis label
    ax.set_ylabel('Net profit in the 100 millions')
    #set title
    plt.title("Company's Profit Over the Past Ten Years")
    #x-axis label
    ax.set_xlabel('years')
    #plotting bar graph
    plt.bar(x_coords, y_coords)
    plt.title('1e8',loc= 'left')

barGraph()
```

```python
def pieChart():
    employee = open("employee_count_by_department.txt", "r")
    line = employee.readline()
    line = employee.readline()
    x_coords = []
    y_coords = []
    while line != '':
        splitline = line.split(',')
        #remove spaces and appends to list
        x_coords.append(splitline[0].strip())
        #remove spaces and appends to list
        y_coords.append(int(splitline[1].strip()))
        line = employee.readline()
    #function to calculate percentages of values
    def absolute_value(val):
        return '{:.1f}%'.format(val, round(val, 1))
    #plotting pie chart
    plt.pie(y_coords, labels = x_coords, autopct = absolute_value)
    #set title
    plt.title('Employees per department')
    plt.show()

pieChart()
```
